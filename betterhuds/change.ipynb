```python
new_recipes = {}
for item in recipes: 
    new_recipes[item["id"]] = {
        "name" : item["name"],
        "vegan" : item["vegan"],
        "vegetarian" : item["vegetarian"],
        "allergens" : item["allergens"],
        "calories" : item["calories"],
        "fiber" : item["dietary_fiber"],
        "sugars" : item["sugars"],
        "fat" : item["total_fat"],
        "carbs" : item["total_carb"],
        "protein" : item["protein"],
        "sodium" : item["sodium"],
        "serving_size" : item["serving_size"],
    }

with open('output.txt', 'w') as f:
    f.write(str(new_recipes))
```
